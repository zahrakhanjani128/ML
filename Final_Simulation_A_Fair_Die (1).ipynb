```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random

num_sim =0
noSim = 1
myarray = []
x_axis = [1, 2, 3, 4, 5 , 6]
y_axis = []

def roll_biased_dice(noSim):
    bias = [0.15, 0.15, 0.15, 0.15, 0.15, 0.25]   # a 6-sided biased die
    np.where(np.random.multinomial(1, bias, size=1)[0]==1)[0][0]+1 # just 1 roll
    
    #If you want to roll the biased die (with the given bias probabilities) for n times, use the following function
    
    def roll(probs, ntimes= noSim):   # roll a len(probs) sided biased die with bias probs for ntimes
        return np.apply_along_axis(lambda x: x.tolist().index(1)+1, 1, 
                              np.random.multinomial(1, bias,  size=6)).tolist()
    myarray = []
    for i in range(int(noSim)): 
        myarray.append(roll(probs=bias, ntimes =noSim))

    print("List of outcomes= ",myarray) 
    visualization(myarray, noSim) 


def visualization(myArray, number):
    for j in range(int(number)):  
        plt.plot(x_axis, myArray[j]) 
    plt.title("Simulate a fair die with a biased 6-sided die")
    plt.xlabel("Dice Sides")
    plt.ylabel("Number of Outcomes")
    plt.show()

    
def main():
    num_sim = input("Enter the Number of Simulation: ")
    roll_biased_dice(num_sim)
        
    
main()    
```
